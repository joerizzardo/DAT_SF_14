(1) Use the file **bank.csv** to explore the dataset. Observe the features: Are they numbers? Are they strings? Are they binary? Are they continuous?


```python
import pandas as pd

cols = ['age','job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome','y']

data = pd.read_csv('C:/Users/cat/DAT_SF_14/homework/hw3/bank.csv', header = True, names = cols, sep =";")

data.groupby(['marital']).marital.count()
```

(2) Learn about label encoders at the following link and use what you learn to transform the features to numerical features: http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

```python
import numpy as np
from sklearn import preprocessing
lab = preprocessing.LabelEncoder()
data.job = lab.fit_transform(data.job)
data.marital = lab.fit_transform(data.marital)
data.education = lab.fit_transform(data.education)
data.default = lab.fit_transform(data.default)
data.balance = lab.fit_transform(data.balance)
data.housing = lab.fit_transform(data.housing)
data.loan = lab.fit_transform(data.loan)
data.contact = lab.fit_transform(data.contact)
data.month = lab.fit_transform(data.month)
data.poutcome = lab.fit_transform(data.poutcome)
data.y = lab.fit_transform(data.y)

data
```

(3) Build a decision tree model to predict whether a prospect will buy the product.

```python
#Defining my columns for splitting my data
cols = data.columns[0:-1]

X = data[cols]
y = data.y

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print X_train
print y_train
print X_test
print y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclass = DecisionTreeClassifier(random_state=1)
treeclass.fit(X_train, y_train)
```

```python
preds = treeclass.predict(X_test)
for i in preds:
    print i
```

(4) Evaluate the accuracy of your decision tree model using cross validation.

```python
#importing cross-validation
from sklearn.cross_validation import cross_val_score

clf = DecisionTreeClassifier(random_state=0)
scores = cross_val_score(clf, X_test, y_test, cv=4)
print scores
```

```python
float(sum(preds == y_test))/ len(y_test)
```
